## CodeT5-220m

CodeT5 is a bit old, requiring some environment setup for compatability.

```python
!pip install transformers==4.41.2
!pip install peft==0.11.1

import os

# Replace processing_class with tokenizer for compatibility with transformers <= 4.44
for script in ['src/train.py', 'src/train_nba.py']:
    with open(script, 'r') as f:
        content = f.read()
    content = content.replace('processing_class=', 'tokenizer=')
    with open(script, 'w') as f:
        f.write(content)
    print(f'Successfully patched {script}')

# Add the NVIDIA nvjitlink library path to LD_LIBRARY_PATH
nvjitlink_path = '/usr/local/lib/python3.12/dist-packages/nvidia/nvjitlink/lib'
os.environ['LD_LIBRARY_PATH'] = f"{nvjitlink_path}:{os.environ.get('LD_LIBRARY_PATH', '')}"

print('Updated LD_LIBRARY_PATH:', os.environ['LD_LIBRARY_PATH'])
```

```python
!python -m src.train --method lora --model Salesforce/codet5p-220m \
    --rank 16 --epochs 3 --lr 1e-4 --batch-size 8
```
